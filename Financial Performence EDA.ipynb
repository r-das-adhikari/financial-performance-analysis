```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import sys
```

```python
def Read_all_csv(folder_path):
    dataframe={}
    for file in glob.glob(os.path.join(folder_path,"*.csv")):
        name = file.split("\\")[-1].replace(".csv","")
        dataframe[name]=pd.read_csv(file)
    return dataframe
```
